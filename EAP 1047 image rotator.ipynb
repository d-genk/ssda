```python
#For some reason all of the EAP 1047 TIFs created by Photoshop were rotated 90 degrees to the left.
#This code will rotate all TIFs in a specified directory 90 degrees to the right.

from PIL import Image
import os
import shutil
```

```python
image_directory = input("Enter absolute path to folder containing images: ")
```

```python
for folder, sub_folders, files in os.walk(image_directory):
    
    print(folder)
    
    for sub_folder in sub_folders:
        if sub_folder == "JPEG":
            shutil.rmtree(folder + '\\' + sub_folder)
    
    for file in files:
        if "tif" in file:
            page = Image.open(folder + '\\' + file)
            
            width = page.size[0]
            height = page.size[1]
            
            if width > height:
                new_page = page.resize((width,width))
                new_page = new_page.rotate(270)
                new_page = new_page.resize((height,width))
                new_page = new_page.save(folder + '\\' + file, format ='TIFF', compression ="tiff_lzw")
            
            page.close()          
```
